### Turkish League Champions

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
titles = {'Fenerbahce':28,
          'Galatasaray':25,
          'Besiktas':20,
                 'Trabzonspor':7,
                 'Bursaspor':1,
                 'Medipol Basaksehir':1}
```

```python
plt.bar(titles.keys(),titles.values())
```

```python
from matplotlib import style
```

```python
style.use('ggplot')
colours = ['darkblue','red', 'black', 'maroon','green','orange']
plt.bar(titles.keys(),titles.values(),color = colours, width = .5, edgecolor = "White")
plt.title('The Number of Team''s Titles')
plt.xlabel('Teams')
plt.ylabel('Titles')
plt.xticks(rotation = 60);
```
